## Erosion and Dilation of images using OpenCV in python

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("input.png")
```

```python
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(img,kernel,iterations = 1)
dilation = cv2.dilate(img, kernel, iterations=1)
```

```python
#cv2.imshow("Original",img)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
#cv2.imshow("Erosion",erosion)
plt.imshow(cv2.cvtColor(erosion, cv2.COLOR_BGR2RGB))
```

```python
#cv2.imshow("Dilation",dilation)
plt.imshow(cv2.cvtColor(dilation, cv2.COLOR_BGR2RGB))
```

## Python | Thresholding techniques using OpenCV | Set-1 (Simple Thresholding)

```python
image=cv2.imread("input1.png")
```

```python
img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# applying different thresholding techniques on the input image all pixels value above 120 will be set to 255
ret,thr1=cv2.threshold(img,120,255,cv2.THRESH_BINARY)
ret,thr2=cv2.threshold(img,120,255,cv2.THRESH_BINARY_INV)
ret,thr3=cv2.threshold(img,120,255,cv2.THRESH_TRUNC)
ret,thr4=cv2.threshold(img,120,255,cv2.THRESH_TOZERO)
ret,thr5=cv2.threshold(img,120,255,cv2.THRESH_TOZERO_INV)

```

```python
# the window showing output images with the corresponding thresholding techniques applied to the input images
cv2.imshow('Binary Threshold', thr1)
cv2.imshow('Binary Threshold Inverted', thr2)
cv2.imshow('Truncated Threshold', thr3)
cv2.imshow('Set to 0', thr4)
cv2.imshow('Set to 0 Inverted', thr5)
   
# De-allocate any associated memory usage 
if cv2.waitKey(0) & 0xff == 27:
    cv2.destroyAllWindows()
```

## Python | Thresholding techniques using OpenCV | Set-2 (Adaptive Thresholding)

```python
im=cv2.imread("book.jpg")
```

```python
img = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
```

```python
thresh1=cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,199,5)
thresh2=cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,199,5)
```

```python
cv2.imshow('Adaptive Mean', thresh1)
cv2.imshow('Adaptive Gaussian', thresh2)
  
     
# De-allocate any associated memory usage  
if cv2.waitKey(0) & 0xff == 27: 
    cv2.destroyAllWindows() 
```

## Python | Thresholding techniques using OpenCV | Set-3 (Otsu Thresholding)

```python
image=cv2.imread("input1.png")
```

```python
img=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
ret, thresh1 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY + 
                                            cv2.THRESH_OTSU)
```

```python
cv2.imshow('Otsu Threshold', thresh1)         
       
# De-allocate any associated memory usage         
if cv2.waitKey(0) & 0xff == 27:
    cv2.destroyAllWindows() 
```
