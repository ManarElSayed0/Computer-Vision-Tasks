## Image Resizing using OpenCV | Python

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread("road.JPG")
```

```python
half = cv2.resize(image, (0, 0), fx = 0.1, fy = 0.1)
bigger = cv2.resize(image, (1050, 1610)) 
stretch_near = cv2.resize(image, (780, 540),interpolation = cv2.INTER_NEAREST)
```

```python
Titles =["Original", "Half", "Bigger", "Interpolation Nearest"]
images =[image, half, bigger, stretch_near]
count = 4
```

```python
for i in range(count):
    plt.subplot(2, 2, i + 1)
    plt.title(Titles[i])
    plt.imshow(images[i])
 
plt.show()
```

## Image Processing in Python (Scaling, Rotating, Shifting and Edge Detection)

## Scaling an Image

```python
try:  
    # Get number of pixel horizontally and vertically.
    (height, width) = image.shape[:2]
  
    # Specify the size of image along with interploation methods.
    res = cv2.resize(image, (int(width / 2), int(height / 2)), interpolation = cv2.INTER_CUBIC)
  
    # Write image back to disk.
    cv2.imwrite('result.jpg', res)
  
except IOError:
    print ('Error while reading files !!!')
```

### Rotating an image

```python
try:
    (rows,cols)=image.shape[:2]
    # getRotationMatrix2D creates a matrix needed for transformation.
    # We want matrix for rotation w.r.t center to 45 degree without scaling.
    M=cv2.getRotationMatrix2D((cols/2,rows/2),45,1)
    res=cv2.warpAffine(image,M,(cols,rows))
      # Write image back to disk.
    cv2.imwrite('result2.jpg', res)
except IOError:
    print ('Error while reading files !!!')
```

### Translating an Image

```python
# Create translation matrix.
# If the shift is (x, y) then matrix would be
# M = [1 0 x]
#     [0 1 y]
# Let's shift by (100, 50).
M = np.float32([[1, 0, 100], [0, 1, 50]])
try:
  
    # Read image from disk.
    
    (rows, cols) = image.shape[:2]
  
    # warpAffine does appropriate shifting given the
    # translation matrix.
    res = cv2.warpAffine(image, M, (cols, rows))
  
    # Write image back to disk.
    cv2.imwrite('result3.jpg', res)
  
except IOError:
    print ('Error while reading files !!!')
```

### Edge detection in an Image

```python
try:
    edges=cv2.Canny(image,100,200)
    cv2.imwrite('result4.jpg', edges)
except IOError:
    print ('Error while reading files !!!')
```

## Python | Image blurring using OpenCV

```python
cv2.imshow("Original_image.JPG",image)
cv2.waitKey(0)
```

```python
#Gaussian Blurr
gaussian=cv2.GaussianBlur(image,(7,7),0)
cv2.imshow("Gaussian_Blur.JPG",gaussian)
cv2.waitKey(0)
```

```python
# Median Blur
median = cv2.medianBlur(image, 5)
cv2.imshow('Median Blurring', median)
cv2.waitKey(0)
```

```python
# Bilateral Blur
bilateral = cv2.bilateralFilter(image, 9, 75, 75)
cv2.imshow('Bilateral Blurring', bilateral)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
