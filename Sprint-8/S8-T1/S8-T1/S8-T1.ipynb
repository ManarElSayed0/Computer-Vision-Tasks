## Introduction to OpenCV

### Reading an image

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image=cv2.imread('road.JPG')
# Extracting the height and width of an image
H,W=image.shape[:2]
# Displaying the height and width
print("Height={} , Width={}".format(H,W))
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

### Extracting the RGB values of a pixel

```python
# Here we have randomly chosen a pixel by passing in 100, 100 for height and width.
B,G,R=image[100,100]
# Displaying the pixel values
print("B={},G={},R={}".format(B,G,R))
```

```python
#We can also pass the channel to extract the value for a specific channel
B=image[100,100,0]
print("B={}".format(B))
```

### Extracting the Region of Interest (ROI)

```python
# We will calculate the region of interest  by slicing the pixels of the image
roi = image[100 : 500, 200 : 700]
plt.imshow(cv2.cvtColor(roi, cv2.COLOR_BGR2RGB))
```

### Resizing the Image

```python
#resize() function takes 2 parameters, the image and the dimensions
resize = cv2.resize(image, (800, 800))
plt.imshow(cv2.cvtColor(resize, cv2.COLOR_BGR2RGB))
```

```python
# Calculating the ratio
ratio = 800 / W

# Creating a tuple containing width and height
dim = (800, int(H* ratio))

# Resizing the image
resize_aspect = cv2.resize(image, dim)
plt.imshow(cv2.cvtColor(resize_aspect, cv2.COLOR_BGR2RGB))
```

### Rotating the Image

```python
# Calculating the center of the image
center=(W//2,H//2)
# Generating a rotation matrix
matrix=cv2.getRotationMatrix2D(center,-45,1.0)
# Performing the affine transformation
rotated=cv2.warpAffine(image, matrix,(W,H))
plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))
```

### Drawing a Rectangle

```python
# We are copying the original image, as it is an in-place operation.
output = image.copy()
  
# Using the rectangle() function to create a rectangle.
#rectangle(Image,Top-left corner co-ordinates,Bottom-right corner co-ordinates,Color (in BGR format),Line width)
rectangle = cv2.rectangle(output, (1500, 900), 
                          (600, 400), (255, 0, 0), 2)
plt.imshow(cv2.cvtColor(rectangle, cv2.COLOR_BGR2RGB))
```

### Displaying text

```python
# Copying the original image
output=image.copy()
# Adding the text using putText() function 
#putText(Image,Text to be displayed,Bottom-left corner co-ordinates from where the text should start,Font,Font size,Color (BGR format),Line width)
text = cv2.putText(output, 'OpenCV Demo', (500, 550), 
                   cv2.FONT_HERSHEY_SIMPLEX, 4, (255, 0, 0), 2)
plt.imshow(cv2.cvtColor(text, cv2.COLOR_BGR2RGB))
```

## Arithmetic Operations on Images using OpenCV | Set-1 (Addition and Subtraction)


```python
import numpy as np
```

```python
img1=cv2.imread("image1.JPG")
img2=cv2.imread("image2.JPG")
```

```python
weightedSum=cv2.addWeighted(img1,0.5,img2,0.4,0)
cv2.imshow('Weighted Image', weightedSum)
# De-allocate any associated memory usage 
if cv2.waitKey(0) & 0xff == 27:
    cv2.destroyAllWindows()
```

```python
image1 = cv2.imread('star.jpg')
image2 = cv2.imread('rhombus.jpg')
```

```python
sub = cv2.subtract(image1, image2)
cv2.imshow('Subtracted Image', sub)
 
# De-allocate any associated memory usage 
if cv2.waitKey(0) & 0xff == 27:
    cv2.destroyAllWindows()
```

## Arithmetic Operations on Images using OpenCV | Set-2 (Bitwise Operations on Binary Images)

```python
img1 = cv2.imread('input1.png') 
img2 = cv2.imread('input2.png')
```

### Bitwise AND operation on Image

```python
dest_and=cv2.bitwise_and(img1,img2,mask=None)
cv2.imshow('Bitwise And', dest_and)
  
# De-allocate any associated memory usage 
if cv2.waitKey(0) & 0xff == 27:
    cv2.destroyAllWindows()
```

### Bitwise OR operation on Image

```python
dest_or = cv2.bitwise_or(img2, img1, mask = None)
cv2.imshow('Bitwise OR', dest_or)
  
# De-allocate any associated memory usage 
if cv2.waitKey(0) & 0xff == 27:
    cv2.destroyAllWindows()
```

### Bitwise XOR operation on Image


```python
dest_xor = cv2.bitwise_xor(img1, img2, mask = None)
cv2.imshow('Bitwise XOR', dest_xor)
  
# De-allocate any associated memory usage 
if cv2.waitKey(0) & 0xff == 27:
    cv2.destroyAllWindows()
```

### Bitwise NOT operation on Image

```python
dest_not1 = cv2.bitwise_not(img1, mask = None)
dest_not2 = cv2.bitwise_not(img2, mask = None)
cv2.imshow('Bitwise NOT on image 1', dest_not1)
cv2.imshow('Bitwise NOT on image 2', dest_not2)
  
# De-allocate any associated memory usage 
if cv2.waitKey(0) & 0xff == 27:
    cv2.destroyAllWindows()
```
